```python
import pandas as pd
import numpy as np

# --- STEP 1: LOAD DATA ---

filename = 'features_chest.csv'
df = pd.read_csv(filename)

print(f"Data Loaded. Shape: {df.shape}")
```

```python
# --- STEP 2: SEPARATE X, y, AND GROUPS ---

# X = All columns EXCEPT the answer ('label') and the identifier ('subject')
# We drop 'subject' because the model shouldn't know WHO the person is, only their signals.
X = df.drop(['label', 'subject'], axis=1)

# y = The answer key (0 = Baseline, 1 = Stress)
y = df['label']

# groups = The list of subjects (used for splitting training/testing data later)
groups = df['subject']

# --- VERIFICATION ---
print("-" * 30)
print(f"X (Features): {X.shape} (Must have 30 columns)")
print(f"y (Labels):   {y.shape} (Must match X rows)")
print(f"Groups:       {groups.unique()}")

# Check if X contains only numbers (Machine Learning only understands numbers)
print("\nData Types in X:")
print(X.dtypes.value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- STEP 3: SPLIT & SCALE ---

# 1. Split Data (70% Train, 30% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Scale Data (CRITICAL for KNN)
# KNN uses distance, so we must normalize features (e.g., Temp vs ACC).
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) # Fit only on training data
X_test_scaled = scaler.transform(X_test)       # Apply same scale to test

print("Setup Complete!")
print(f"Training Data: {X_train.shape[0]} rows")
print(f"Test Data:     {X_test.shape[0]} rows")
```

```python
from sklearn.linear_model import LogisticRegression
import time

# --- STEP 4: TRAIN LOGISTIC REGRESSION MODEL ---
print("--> Training Logistic Regression Model...")
start_time = time.time()

# We use max_iter=1000 to ensure the solver finds the optimal solution
# Logistic Regression requires scaled data (X_train_scaled) to work properly
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Fitting the model to the training data
log_reg.fit(X_train_scaled, y_train)

print(f"Training completed in {time.time() - start_time:.4f} seconds.")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# --- STEP 5: TEST AND EVALUATE LOGISTIC REGRESSION MODEL---
# Generating predictions on the test set
y_pred_log = log_reg.predict(X_test_scaled)

# Calculating Accuracy
log_acc = accuracy_score(y_test, y_pred_log)

print(f"Logistic Regression Test Accuracy: {log_acc:.4f} ({log_acc*100:.2f}%)")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_log, target_names=['Baseline', 'Stress'], digits=4))
```

```python
from sklearn.neighbors import KNeighborsClassifier
import time

# --- STEP 6: TRAIN KNN MODEL---

print("--> Training KNN (this might take a few seconds)...")
start_time = time.time()

# We use the SCALED data for KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

print(f"KNN Training finished in {time.time() - start_time:.2f} seconds.")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# --- STEP 7: EVALUATE KNN MODEL ---

# Predict using the SCALED test set
y_pred_knn = knn.predict(X_test_scaled)

# Calculate Accuracy
acc_knn = accuracy_score(y_test, y_pred_knn)

print(f"KNN Accuracy: {acc_knn:.4f} ({acc_knn*100:.2f}%)")
print("\nDetailed Report:")
print(classification_report(y_test, y_pred_knn, digits=4))
```

```python
from sklearn.ensemble import RandomForestClassifier

# --- STEP 8: TRAIN RANDOM FOREST ---

print("--> Training Random Forest...")
start_time = time.time()

# Random Forest does NOT need scaling. We use raw X_train.
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

print(f"Random Forest Training finished in {time.time() - start_time:.2f} seconds.")
```

```python
# --- STEP 9: EVALUATE RANDOM FOREST ---

# Predict using the ORIGINAL test set (X_test)
y_pred_rf = rf.predict(X_test)

# Calculate Accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"Random Forest Accuracy: {acc_rf:.4f} ({acc_rf*100:.2f}%)")
print("\nDetailed Report:")
print(classification_report(y_test, y_pred_rf, digits=4))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
import time

# --- STEP 10: LEAVE-ONE-SUBJECT-OUT (LOSO) VALIDATION FOR KNN ---

# Initialize lists to store results for each subject
accuracies_knn_raw = []
f1_scores_knn_raw = []

# Get the list of all unique subjects
subject_list = groups.unique()

print(f"Starting LOSO Validation for KNN on {len(subject_list)} subjects (Global Scaling)...")
print("-" * 50)

for sub in subject_list:
    # 1. SPLIT: Use X_scaled (from Step 2) for KNN
    test_mask = (groups == sub)
    train_mask = ~test_mask
    
    # We must use the globally scaled data created in Step 2
    # Convert X_train_scaled/X_test_scaled back to a temporary split for LOSO
    # Note: For LOSO, we re-scale based on the specific train/test split of that fold
    scaler_loso = StandardScaler()
    X_train_loso = scaler_loso.fit_transform(X[train_mask])
    X_test_loso = scaler_loso.transform(X[test_mask])
    
    y_train_loso = y[train_mask]
    y_test_loso = y[test_mask]
    
    # 2. TRAIN KNN
    knn_loso = KNeighborsClassifier(n_neighbors=5)
    knn_loso.fit(X_train_loso, y_train_loso)
    
    # 3. PREDICT
    y_pred = knn_loso.predict(X_test_loso)
    
    # 4. EVALUATE
    acc = accuracy_score(y_test_loso, y_pred)
    f1 = f1_score(y_test_loso, y_pred, average='weighted')
    
    accuracies_knn_raw.append(acc)
    f1_scores_knn_raw.append(f1)
    
    print(f"Tested on {sub}: Accuracy = {acc:.4f}")

# --- FINAL RESULTS FOR KNN ---
print("-" * 50)
print(f"KNN Average Accuracy (Global Scaling): {np.mean(accuracies_knn_raw):.4f}")
print(f"KNN Average F1-Score (Global Scaling): {np.mean(f1_scores_knn_raw):.4f}")
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# --- STEP 11: LEAVE-ONE-SUBJECT-OUT (LOSO) VALIDATION FOR RANDOM FOREST ---

# Initialize lists to store results for each subject
accuracies = []
f1_scores = []

# Get the list of all unique subjects
subject_list = groups.unique()

print(f"Starting LOSO Validation on {len(subject_list)} subjects...")
print("-" * 50)

for sub in subject_list:
    # 1. SPLIT: Define Train and Test based on Subject ID
    # Test set = The current subject (e.g., 'S2')
    # Train set = Everyone else
    
    # We create a mask (True/False list) to select rows
    test_mask = (groups == sub)
    train_mask = ~test_mask # The symbol '~' means NOT (everyone else)
    
    X_train_loso = X[train_mask]
    y_train_loso = y[train_mask]
    
    X_test_loso = X[test_mask]
    y_test_loso = y[test_mask]
    
    # 2. TRAIN: Random Forest (The winner of our previous test)
    rf_loso = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
    rf_loso.fit(X_train_loso, y_train_loso)
    
    # 3. PREDICT
    y_pred = rf_loso.predict(X_test_loso)
    
    # 4. EVALUATE
    acc = accuracy_score(y_test_loso, y_pred)
    f1 = f1_score(y_test_loso, y_pred, average='weighted')
    
    accuracies.append(acc)
    f1_scores.append(f1)
    
    print(f"Tested on {sub}: Accuracy = {acc:.4f}")

# --- FINAL RESULTS ---
print("-" * 50)
print(f"Average Accuracy: {np.mean(accuracies):.4f} ({np.mean(accuracies)*100:.2f}%)")
print(f"Average F1-Score: {np.mean(f1_scores):.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# --- STEP 12: SUBJECT-SPECIFIC NORMALIZATION ---

# Goal: Remove individual physiological bias.
# For example, if Subject A has a resting heart rate of 60 and Subject B has 90,
# the model might get confused. We standardize each subject to their own baseline.

print("--> Applying Subject-Specific Normalization...")

# Create a copy of X to hold the normalized data
# We use copy() to ensure we don't modify the original X
X_norm = X.copy()

# Iterate over each unique subject (S2, S3, etc.)
for sub in groups.unique():
    # Create a mask to select rows for the current subject
    mask = (groups == sub)
    
    # Isolate the data for this specific subject
    sub_data = X.loc[mask]
    
    # Initialize a scaler specifically for this subject
    scaler = StandardScaler()
    
    # Fit the scaler on this subject's data and transform it
    # This converts absolute values into Z-scores (relative values)
    X_norm.loc[mask] = scaler.fit_transform(sub_data)

print("Normalization Complete! 'X_norm' is now ready for training.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# --- STEP 13: LOSO VALIDATION (USING SUBJECT-SPECIFIC NORMALIZATION) FOR RANDOM FOREST---

accuracies_norm = []
f1_scores_norm = []
subject_list = groups.unique()

print(f"Starting LOSO Validation on {len(subject_list)} subjects (Normalized Data)...")
print("-" * 50)

for sub in subject_list:
    # 1. SPLIT DATA
    # We use the 'groups' column to split, but we pull data from 'X_norm'
    test_mask = (groups == sub)
    train_mask = ~test_mask
    
    X_train_loso = X_norm[train_mask]
    y_train_loso = y[train_mask]
    
    X_test_loso = X_norm[test_mask]
    y_test_loso = y[test_mask]
    
    # 2. TRAIN MODEL
    # We use RandomForest with 50 trees (fast and effective)
    rf_loso = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
    rf_loso.fit(X_train_loso, y_train_loso)
    
    # 3. PREDICT
    y_pred = rf_loso.predict(X_test_loso)
    
    # 4. EVALUATE
    acc = accuracy_score(y_test_loso, y_pred)
    f1 = f1_score(y_test_loso, y_pred, average='weighted')
    
    accuracies_norm.append(acc)
    f1_scores_norm.append(f1)
    
    print(f"Tested on {sub}: Accuracy = {acc:.4f}")

# --- FINAL RESULTS ---
print("-" * 50)
mean_acc = np.mean(accuracies_norm)
mean_f1 = np.mean(f1_scores_norm)

print(f"Average Accuracy (Normalized): {mean_acc:.4f} ({mean_acc*100:.2f}%)")
print(f"Average F1-Score (Normalized): {mean_f1:.4f}")

# Compare with previous baseline (~67.94%)
improvement = (mean_acc - 0.6794) * 100
print(f"\nImprovement over baseline: +{improvement:.2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# --- STEP 14: LOSO VALIDATION (USING SUBJECT-SPECIFIC NORMALIZATION) FOR KNN ---

accuracies_knn = []
f1_scores_knn = []
subject_list = groups.unique()

print(f"Starting LOSO Validation for KNN on {len(subject_list)} subjects...")
print("-" * 50)

for sub in subject_list:
    # 1. SPLIT DATA (Using X_norm)
    test_mask = (groups == sub)
    train_mask = ~test_mask
    
    X_train_loso = X_norm[train_mask]
    y_train_loso = y[train_mask]
    
    X_test_loso = X_norm[test_mask]
    y_test_loso = y[test_mask]
    
    # 2. TRAIN KNN MODEL
    # We use K=5 (default)
    knn_loso = KNeighborsClassifier(n_neighbors=5)
    knn_loso.fit(X_train_loso, y_train_loso)
    
    # 3. PREDICT
    y_pred = knn_loso.predict(X_test_loso)
    
    # 4. EVALUATE
    acc = accuracy_score(y_test_loso, y_pred)
    f1 = f1_score(y_test_loso, y_pred, average='weighted')
    
    accuracies_knn.append(acc)
    f1_scores_knn.append(f1)
    
    print(f"Tested on {sub}: Accuracy = {acc:.4f}")

# --- FINAL RESULTS FOR KNN ---
print("-" * 50)
mean_acc_knn = np.mean(accuracies_knn)
mean_f1_knn = np.mean(f1_scores_knn)

print(f"KNN Average Accuracy: {mean_acc_knn:.4f} ({mean_acc_knn*100:.2f}%)")

# Comparison with Random Forest (assuming 95.83% from previous step)
diff = mean_acc_knn - 0.9583
if diff > 0:
    print(f"Result: KNN won by {diff*100:.2f}%! ðŸ†")
else:
    print(f"Result: Random Forest is still better by {abs(diff)*100:.2f}%.")
```

```python
from sklearn.model_selection import GridSearchCV, LeaveOneGroupOut
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import time

# --- STEP 15: HYPERPARAMETER TUNING (GRID SEARCH) ---
# Goal: Find the mathematically optimal configuration for the Random Forest.
# Critical: We must use LeaveOneGroupOut (LOSO) inside the GridSearch to avoid data leakage.

print("--> Starting Grid Search (Optimization)...")
print("    Note: This is computationally heavy. It might take 5-10 minutes.")

# 1. DEFINE THE GRID
# These are the settings we want to compete against each other.
# We keep it small for demonstration purposes.
param_grid = {
    'n_estimators': [50, 100],        # Number of trees
    'max_depth': [None, 10],          # Maximum depth of each tree (None = infinite)
    'min_samples_leaf': [1, 4]        # Minimum samples required at a leaf node
}

# 2. SETUP THE SPLITTER
# This tells GridSearch: "Never mix subjects! Test on one subject at a time."
loso_cv = LeaveOneGroupOut()

# 3. CONFIGURE GRID SEARCH
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=loso_cv,               # Use LOSO validation
    scoring='accuracy',       # Maximize Accuracy
    verbose=1,                # Show progress
    n_jobs=-1                 # Use all CPU cores
)

# 4. EXECUTE (FIT)
# Important: We must pass 'groups' so the splitter knows which rows belong to whom.
start_time = time.time()
grid_search.fit(X_norm, y, groups=groups)
end_time = time.time()

# --- RESULTS ---
print("-" * 50)
print(f"Optimization finished in {(end_time - start_time)/60:.1f} minutes.")
print(f"Best Parameters Found: {grid_search.best_params_}")
print(f"Best Accuracy (Mean LOSO): {grid_search.best_score_:.4f} ({grid_search.best_score_*100:.2f}%)")

# Compare with our previous manual result
print("-" * 50)
print("Interpretation:")
if grid_search.best_score_ > 0.9583:
    print("âœ… The tuning squeezed out a bit more performance!")
else:
    print("ðŸ†— The default parameters were already excellent. This proves robustness.")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# --- STEP 16: VISUALIZATION - GLOBAL CONFUSION MATRIX FOR RANDOM FOREST ---

y_true_all = []
y_pred_all = []
subject_list = groups.unique()

print("--> Generating Global Confusion Matrix (Random Forest)...")

for sub in subject_list:
    # 1. SPLIT DATA (Using Normalized Data 'X_norm')
    test_mask = (groups == sub)
    train_mask = ~test_mask
    
    X_train = X_norm[train_mask]
    y_train = y[train_mask]
    
    X_test = X_norm[test_mask]
    y_test = y[test_mask]
    
    # 2. TRAIN MODEL
    rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    # 3. PREDICT
    pred = rf.predict(X_test)
    
    # 4. STORE RESULTS
    y_true_all.extend(y_test)
    y_pred_all.extend(pred)

# --- PLOT THE NORMALIZED MATRIX ---

# 1. Calculate the absolute confusion matrix
cm = confusion_matrix(y_true_all, y_pred_all)

# 2. Normalize the matrix (divide each row by its sum to get percentages)
# axis=1 ensures we divide by the total of "True Labels"
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]



# 3. Plot using Seaborn
plt.figure(figsize=(8, 6))
# annot=True: shows the numbers
# fmt='.2%': formats as percentage with 2 decimal places
sns.heatmap(cm_percent, annot=True, fmt='.2%', cmap='Blues', cbar=True,
            xticklabels=['Baseline (0)', 'Stress (1)'],
            yticklabels=['Baseline (0)', 'Stress (1)'])

plt.title('Global Confusion Matrix - Percentages (LOSO Validation)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# --- PRINT DETAILED METRICS ---
print("\nFinal Classification Report:")
print(classification_report(y_true_all, y_pred_all, target_names=['Baseline', 'Stress'], digits=4))
```
